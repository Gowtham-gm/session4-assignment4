```python
#Write a function to find moving average in an array over a window:
#Test it over [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150] and window of 3.

def movingaverage(data,n):
    movingaverage=[]
    m=1
    for i in data:
        if m<n:
            a=None
        else:
            a=sum(data[m-n:m])/n
    
            movingaverage.append(a)
        m+=1
    return(movingaverage)

data=[3,5,7,2,8,10,11,65,72,81,99,100,150]
n=3

g=movingaverage(data,n)
print('the moving average of the given data is :')
print(g)
```

```python
#How-to-count-distance-to-the-previous-zero
#For each value, count the difference back to the previous zero (or the start of the Series,
#whichever is closer)
#create a new column 'Y'
#Consider a DataFrame df where there is an integer column 'X'


import pandas as pd
df = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})

X_flatdata = df.index[df['X']].tolist()  

prev_zero = [(X_flatdata[i::-1]+[0]).index(0) for i in range(len(X_flatdata))] #to count distance to previous data

df['Y'] = prev_zero  #adding Y column to the existing dataframe
df
        
 
        
```

```python
#create  DatetimeIndex that contains each business day of 2015 and use it to index a Series of
#random numbers

import pandas as pd
import numpy as np
import datetime
datetimeindex=pd.date_range(start='2015-01-01',end='2015-12-31',freq='B');datetimeindex
s=pd.DataFrame(np.random.rand(len(datetimeindex)),index=datetimeindex) #setting daterange as index to the dataframe
s
```

```python
#Find the sum of the values in s for every Wednesday
print('The sum of the values in s for every wednesday is:')
s[s.index.weekday==2].sum()
```

```python
#Average For each calendar month
print('The average of the values for each month is:')
s.resample('M').mean()
```

```python
#For each group of four consecutive calendar months in s, find the date on which the highest
#value occurred.
highestin_4months=s.groupby(pd.Grouper(freq='4M')).idxmax()
print('the highest number in the 4 cosecutive months are:')
highestin_4months
```
